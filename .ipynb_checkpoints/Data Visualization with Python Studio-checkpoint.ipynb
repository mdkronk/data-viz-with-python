```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", error_bad_lines=False)
df.head()
```

```python
years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years
```

```python
df.head(2)
```

```python
df.info()
```

```python
#look for any columns that have all the same value
df.nunique()
```

```python
# check for duplicates
df.loc[df.duplicated(), :]
```

```python
cleaned_data = df.drop(['language_code','ratings_count','text_reviews_count', 'average_rating' ], axis=1)
cleaned_data.head()
```

```python
df_years = cleaned_data['year']
print(df_years.max())
print(df_years.min())
```

```python
cleaned_data['decade'] = cleaned_data.year.apply(lambda x: str(x)[:3]+'0')
cleaned_data.head()
```

```python
df_grouped = cleaned_data.groupby('decade').size()
df_grouped
```

With your dataframe set up, time to start creating visualizations!

```python
# Use your Python skills to create a visualization here.
import mplcyberpunk
style.use('cyberpunk')

plt.figure(figsize = (14,6))
graph = sb.lineplot(y = df_grouped, x = df_grouped.index)
plt.xticks(rotation = 90, color = 'white', fontsize = 14, fontweight = 'bold')
plt.yticks(color = 'white', fontsize = 14, fontweight = 'bold')
plt.title('Books Published By Decade From 1900 To 2020',color = 'white', fontsize = 20, fontweight = 'bold')
plt.xlabel('Decade',color = 'white', fontsize = 16, fontweight = 'bold')
plt.ylabel('Books Published',color = 'white',fontsize = 16, fontweight = 'bold')
plt.show()
```
